```python
import numpy as np
import glob
from PIL import Image
import cv2 as cv
```

```python
img_rows = 224
img_cols = 224
channel = 3
filelistTrain = glob.glob('../data/*')
```

```python
def load_img(i):

    if '_g_' in filelistTrain[i]:
        label = 1
    else:
        label = 0

    #print(self.filelistTrain[i])
    img = Image.open(filelistTrain[i]).convert('RGB')
    img = img.resize([img_rows, img_cols])
    img = np.array(img)
    img = np.array(cv.normalize(img.astype(np.float32), np.zeros((img_rows, img_cols, channel), dtype=np.float32), -1, 1, cv.NORM_MINMAX))

    return img, label
```

```python
def data2predict(batch_size):
    
    ctrl = 0
    
    print(ctrl)
    
    while ctrl < 2358:

        batch_imgs = []

        ini = np.random.choice(len(filelistTrain) - batch_size, 1)[0]

        for i in range(ini, ini + batch_size):

            image_sample, _ = load_img(i)
            print(filelistTrain[i])
            batch_imgs.append(image_sample)

        batch_imgs = np.array(batch_imgs).astype(np.float32)
        
        ctrl += batch_size
        

        yield batch_imgs
```

```python
data = data2predict(batch_size=32)
see = 0 
for i in data:
    print(see)
    print(i.shape)
    see += 1
```
