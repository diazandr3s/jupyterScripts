```python
import glob
import cv2
import matplotlib.pyplot as plt
from skimage.morphology import label
from skimage.measure import regionprops
import numpy as np
```

```python
# Extracting file names
nameIdx = 18
folderResults = './croppedMESSIDOR/'


allfiles = glob.glob("./MESSIDOR_YES/OD/*")
print(allfiles[0][nameIdx:-4])
```

```python
def extractOD(od):
    maxiOD = np.max(od) 
    #print(maxiOD)
    
    # Bounding box. We first need to create a label image
    label_img = label(od[:,:,1] > maxiOD/2 )   
    props = regionprops(label_img)  
    #print(type(props))
    #print(props[0].bbox)
    
    #print('Centroid:' + str(props[0].centroid))
    
    newR = 1.5*(props[0].equivalent_diameter/2)
    
    minRow = props[0].centroid[0] - newR
    maxRow = props[0].centroid[0] + newR
    minCol = props[0].centroid[1] - newR
    maxCol = props[0].centroid[1] + newR
    
    
    return int(minRow),int(maxRow),int(minCol),int(maxCol),(props[0].equivalent_diameter/2)*0.5
```

```python
for i in range(0,len(allfiles)):
    
    #print(len(allfiles))
    
    od = cv2.imread(allfiles[i]) 
    
    # Extracting bounding box
    minRow,maxRow,minCol,maxCol,newR = extractOD(od)
      
    print(od.shape)
    
    # plotting OD    
    #newOD = cv2.cvtColor(od, cv2.COLOR_RGB2BGR)
    #plt.imshow(newOD)
    #plt.show()
    
    im = cv2.imread('./MESSIDOR_YES/Originals/' + allfiles[i][nameIdx:])
    
    # Cropping image
    imCrop = im[minRow:maxRow,minCol:maxCol,:]
    print(imCrop.shape)
    print(imCrop.size)
    
    nonZeros = np.count_nonzero(imCrop)
    zerosVal = imCrop.size - nonZeros
    
    perZeros = zerosVal/imCrop.size
    
    print('Non zeros: ' + str(nonZeros))
    print('Zeros: ' + str(zerosVal))    
    
    print('Percentage zeros: ' + str(perZeros))
    
      
    
    newIm = cv2.cvtColor(imCrop, cv2.COLOR_RGB2BGR)
    #plt.imshow(newIm)
    #plt.show()
    
    # If the percentage of zeros is bigger than 0.009 we change the cropping
    if perZeros > 0.009:
        imCrop = im[minRow+int(round(newR)):maxRow-int(round(newR)),minCol+int(round(newR)):maxCol-int(round(newR)),:]
        newIm = cv2.cvtColor(imCrop, cv2.COLOR_RGB2BGR)
        #plt.imshow(newIm)
        #plt.show()
        
    
    # Saving cropped images
    cv2.imwrite(folderResults + allfiles[i][nameIdx:-4] + '.png', imCrop)
    
    
```
